```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

# Step 1: Load your data
df = pd.read_csv("../data/feature_data.csv")

# Step 2: Load pretrained BERT model and tokenizer
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Step 3: Create sentiment classifier pipeline
classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer, device=-1)

# Step 4: Apply the classifier to each review
df['sentiment'] = df['feedback'].apply(lambda x: classifier(x)[0]['label'])
df['confidence'] = df['feedback'].apply(lambda x: classifier(x)[0]['score'])

# Step 5: View results
print(df)
```
